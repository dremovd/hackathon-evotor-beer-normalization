# Пример работы API построенного на модели линейной классификации которая использует только текст

```python
import pandas as pd
```

```python
# Исключены несколько строк у которых нарушен формат
# Кодировка utf-8
data = pd.read_csv('../MASTER_DATA_beer_hackaton_fixed.csv.gz')
```

Метод API:

https://qr-kassa.ru:9876/name

Ожидает в теле запрос JSON: {'name': name}

```python
import requests
import tqdm

results = []
for name in tqdm.tqdm_notebook(data['o.item_name'].sample(50).values):
    response = requests.post('https://qr-kassa.ru:9876/name', json={'name': name}).json()
    results.append(response)

results = pd.DataFrame(results)
```

```python
targets = [
    'Алкоголь', 'Объем', 'Производитель', 'Бренд',
    'Саб-бренд', 'Тип упаковки', 'Мультипак', 'тип'
]
```
